```python
import matplotlib.pyplot as plt
```

```python
x= [50,42,39,34,32,48,22,17,8,13]
y= [17,19,21,22,24,23,34,31,43,35]
```

```python
plt.scatter(x,y)
plt.xlabel('Total Number of Rides (per City)')
plt.ylabel('Average Fares ($)')
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#There are two ways to create scatter plots plt.plot & plt.scatter

plt.plot(y_axis, x_axis, 'o', color = "red", label = "Chicago" )
#plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.legend()
plt.xlabel("Fares ($)")
plt.ylabel("Date")
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
```

```python
# Create a Bubble Chart using the Matlab method
# s = y_axis is the fuction that determin the size of each marker(Bubble)

#plt.scatter(x_axis, y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)

#declared a variable and created an empty list, used for loop statemnet for data in y_axis multiply
#-it by 3
```

```python
#Create a Bubble Chart Using the Object-Oriented Interface

#Using object oriented method
#Change the color of the markers to sky blue.

#Change the size of the markers to 5 times each data point.

#Make the color 20% transparent.

#Add a black edge color to the circles.

#Make the linewidth of the circles 2 points.

#Add a legend for the city of Boston.

#Add a title.

#Switch the axis so that the Fare($) data is on the x-axis.

#Add a limit for the x-axis data.

#nvert the y-axis so that January is at the top.


fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)

ax.scatter(x_axis, y_axis, s=y_axis_larger, color = "skyblue")
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

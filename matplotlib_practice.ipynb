```python
# Import dependencies. matplotlib.pyplot = from matplotlib import pyplot

import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
#To create a figure, use fig, ax = plt.subplots().
#To define the axes, use ax.plot().
#Inside the parentheses of ax.plot(), add the axes separated by a comma.
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.

#plt.plot(x_axis, y_axis, label='Boston')

#plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

#plt.xlim("Jan", "June") to specify x-axis
#plt.xlim(0, 5)

#plt.figure(figsize=(20, 10))
#plt.subplots(figsize=(20, 10))

# Create a title.

plt.title("PyBer Fare by Month")

# Add a grid.

plt.grid()

# Add the legend.

plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.

plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Create the plot - horizontal bar

plt.barh(x_axis, y_axis)
```

```python
# Create the plot - inverted this is y-axis, x-axis

plt.barh(y_axis, x_axis)
```

```python
# Create the plot. Invert the month to show in proper order Jan-Dec 

plt.barh(x_axis, y_axis, color = "magenta", label ='Boston')
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
```
